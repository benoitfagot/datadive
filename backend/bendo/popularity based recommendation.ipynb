```python
def create_file(userID,city):
    df_business = pd.read_json("../dataset/business.json", lines=True)
    df = pd.read_json("../dataset/cleaned/review_stars_"+city+"_2016_2019.json", lines=True)
    Sum = df.groupby(by="business_id",as_index=False)['stars'].sum()
    Sum.rename(columns={'stars': 'business_popularity'}, inplace=True)
    df = pd.merge(Sum,df_business,how="inner", on='business_id')
    df = df[df.is_open == 1]
    df = df.sort_values(by='business_popularity', ascending=False)
    df.to_json("../dataset/cleaned/business_popularities_"+city+"_2016_2019.json", orient='records', lines=True)

#create_file("userID","Mesa")

```
